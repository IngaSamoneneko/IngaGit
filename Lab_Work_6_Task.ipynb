```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns


import pandas as pd
xl = pd.ExcelFile('D:\\Python11\Lab Work 6\FOR_LAB_6.xlsx')
#xl.sheet_names
```

```python
Orders=xl.parse('Orders')
Customers=xl.parse('Customers')
Zip=xl.parse('Zip')
Product=xl.parse('Product')
Returns=xl.parse('Returns')
#Orders.head()
```

```python
df = pd.DataFrame(Orders)
xl2=df.groupby(by=['Ship Mode'])['Shipping Cost'].sum()

a=(xl2[0],xl2[1],xl2[2])
b=(xl2.index[0],xl2.index[1],xl2.index[2])


fig, ax = plt.subplots()
ax.pie(a, labels=b, autopct='%1.1f%%',pctdistance=1.25, labeldistance=.6)

plt.title("Shipping Cost by Ship Mode")
plt.show()
```

```python
xl3=df.groupby(by=['Customer ID'])['Unit Price'].sum().sort_values(ascending=False)
xl3.head()

#total=[]
#price=[]
#for i in range(len(xl3)):
#     total+=[xl3.index[i]]
#     price+=[round(xl3.loc[xl3.index[i]],2)]
#     i+=1
        

data={}        
for i in range(5):
     d=str(xl3.index[i])
     data[d]=round(xl3.loc[xl3.index[i]],2)
     i+=1
    
courses = list(data.keys())
values = list(data.values())

plt.barh(courses, values, color ='maroon')
plt.xlabel("Total Sales")
plt.ylabel("Customer ID")
plt.title("Sales by Customers")
plt.show()
```
